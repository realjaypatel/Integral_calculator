```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 + 3*x + 2

# Define the derivative of the function
def f_prime(x):
    return 2*x + 3

# Define the point at which you want to draw the tangent
z = 2

# Calculate the slope of the tangent line at point z
slope = f_prime(z)

# Define the tangent line using point-slope form (y - y1 = m(x - x1))
# We use the point (z, f(z)) and the slope calculated above
def tangent_line(x):
    return slope * (x - z) + f(z)

# Generate x values for plotting
x_values = np.linspace(z - 2, z + 2, 400)
y_values = f(x_values)

# Generate y values for the tangent line
tangent_y_values = tangent_line(x_values)

# Plot the function and the tangent line
plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, tangent_y_values, label='Tangent at z')

# Plot the point where the tangent is drawn
plt.scatter(z, f(z), color='red', label='Point z')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Tangent Line')
plt.legend()
plt.grid(True)
plt.show()
```
